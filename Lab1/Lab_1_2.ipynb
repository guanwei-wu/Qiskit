# 2. (a) (4 points)

```python
import qiskit
import numpy as np
from qiskit import *
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from math import sqrt, pi

import math
from qiskit.tools.visualization import plot_histogram
```

```python
q = QuantumRegister(2)
c = ClassicalRegister(2)
```

```python
circuit = QuantumCircuit(q, c)
circuit.h(q[0])
circuit.h(q[1])
circuit.draw()
```

```python
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
circuit.measure(q, c)
circuit.draw()
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(circuit, simulator, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```

```python
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(circuit, simulator, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```

```python
circuit = QuantumCircuit(q, c)
circuit.h(q[0])
circuit.ry(-2*math.acos(1/3**(1/2)), q[0])

# ===== {|+⟩, |−⟩} basis =====
circuit.h(q[0])
# ==========================

# ===== swap =====
circuit.swap(q[0],q[1])
# ================

circuit.draw()
```

```python
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
circuit = QuantumCircuit(q, c)
circuit.h(q[0])
circuit.ry(-2*math.acos(1/3**(1/2)), q[0])

# ===== {|+⟩, |−⟩} basis =====
circuit.h(q[0])
# ==========================

# ===== swap =====
circuit.swap(q[0],q[1])
# ================

# ===== equivalent swap =====
circuit.cx(q[0],q[1])
circuit.cx(q[1],q[0])
circuit.cx(q[0],q[1])
# ================

circuit.draw()
```

```python
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit, simulator)
result = job.result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
circuit.measure(q, c)
circuit.draw()
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(circuit, simulator, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```
