# 3. (a) (4 points)

```python
import qiskit
import numpy as np
from qiskit import *
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from math import sqrt, pi

import math
from qiskit.tools.visualization import plot_histogram
```

```python
q_a = QuantumRegister(1)
c_a = ClassicalRegister(1)
q_b = QuantumRegister(1)
c_b = ClassicalRegister(1)
```

```python
circuit_a = QuantumCircuit(q_a, c_a)
circuit_b = QuantumCircuit(q_b, c_b)
```

```python
circuit_a.h(q_a[0])

# ===== another input =====
# circuit_a.ry(-2*math.acos(1/3**(1/2)), q_a[0])
# =========================

circuit_a.x(q_a[0])

circuit_b.h(q_b[0])

# ===== another input =====
# circuit_b.ry(-2*math.acos(1/3**(1/2)), q_b[0])
# =========================

circuit_b.y(q_b[0])
circuit_b.z(q_b[0])
```

```python
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit_a, simulator)
result = job.result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
simulator = Aer.get_backend('statevector_simulator')
job = execute(circuit_b, simulator)
result = job.result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
circuit_a.measure(q_a, c_a)
circuit_a.draw()
```

```python
circuit_b.measure(q_b, c_b)
circuit_b.draw()
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(circuit_a, simulator, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(circuit_b, simulator, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```
